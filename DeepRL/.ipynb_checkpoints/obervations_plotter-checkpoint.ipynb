```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

data = pickle.load(open("observations.pkl", "rb"))
```

```python
num_rl = 2
num_steps = 1500
dt = 0.1

vel_lead = [a[num_rl] * 30 for a in data]
vel1 = [a[num_rl +1] * 30 for a in data]
vel2 = [a[num_rl +2] * 30 for a in data]

# plot the speed profile
plt.figure(figsize=(14,9))
plt.title("Velocity Profile of Vehicles in the Platoon", fontsize=25)
plt.xlabel("time (s)", fontsize=20)
plt.ylabel("velocity (m/s)", fontsize=20)
plt.plot(np.arange(num_steps)*dt, vel_lead[:num_steps], linewidth=2)
plt.plot(np.arange(num_steps)*dt, vel1[:num_steps], linewidth=2)
plt.plot(np.arange(num_steps)*dt, vel2[:num_steps], linewidth=2)
plt.legend(["leader", "AV1", "AV2"], fontsize=20)
plt.show()
```
